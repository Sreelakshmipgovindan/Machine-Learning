***Bivariate analysis***

```python
Malignant_cancer=df[df.diagnosis==1]
Benign_cancer=df[df.diagnosis==0]
```

```python
def biplot(data, var,label_rotation):
    fig, (ax1, ax2) = plot.subplots(ncols=2, figsize=(20,10))
    r1=sbn.barplot(ax=ax1,x=Malignant_cancer[var].value_counts().index, data=Malignant_cancer, y= 100.* Malignant_cancer[var].value_counts(normalize=True))
    ## condition is applied for 1st bar plot
    if(label_rotation):
        r1.set_xticklabels(r1.get_xticklabels(),rotation=90,fontsize=14) ## setting x stick labels for 1st bar plot
    ## providing title and x and y axis
    ax1.set_title('Visualisation of '+ '%s' %var +' - Malignant', fontsize=17) ## setting title of 1st bar plot
    ax1.set_xlabel('%s' %var,fontsize=16) ## setting x labels
    ax1.set_ylabel("Cancer Percentage",fontsize=16) ## setting y-label for Malignant
    ## Bar plot
    r2=sbn.barplot(ax=ax2,x=Benign_cancer[var].value_counts().index, data=Benign_cancer, y= 100.* Benign_cancer[var].value_counts(normalize=True))    ## condition is applied for 2nd bar plot
    if(label_rotation):
        r2.set_xticklabels(r2.get_xticklabels(),rotation=90,fontsize=14) ## setting x stick labels for 2nd bar plot
    ## setting name of axis
    ax2.set_xlabel('%s' %var, fontsize=16) ## setting x labels
    ax2.set_ylabel("Cancer Percentage", fontsize=16)  ## setting y-label for benign
    ax2.set_title('Visualisation '+ '%s' %var +' - Benign', fontsize=16)  ## setting title of 1st bar plot
    ## Displaying plot
    plot.show() ## displaying plots
```

```python
biplot(df, 'radius_mean',False)
```

```python
# displaying Correlation map
f,ax = plot.subplots(figsize=(18, 18))
sbn.heatmap(df.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax);
```

```python
def bivariateplots(var1,var2): ## defining two variables
    fig, (ax1,ax2) = plot.subplots(ncols=2,figsize=(18,9))
    sbn.scatterplot(ax=ax1,x=Malignant_cancer[var1],y=Malignant_cancer[var2],color='green') ## applying conditions for scatter plot of malignant cancer
    ax1.set_xlabel(var1) ## setting 1st variable as the x labels
    ax1.set_ylabel(var2) ## setting 2nd variable as the y labels
    ax1.set_title("Malignant cancer") # setting title for malignant cancer
    ##Scatter plot
    sbn.scatterplot(ax=ax2,x=Benign_cancer[var1],y=Benign_cancer[var2],color='magenta') ## applying conditions for scatter plot of benign cancer
    ##Title
    ax2.set_xlabel(var1) ## setting 1st variable as the x labels (benign cancer)
    ax2.set_ylabel(var2) ## setting 2nd variable as the y labels (benign cancer)
    ax2.set_title("Benign cancer") # setting title for belign cancer
    ## displaying plot
    plot.show()
```

```python
bivariateplots('smoothness_mean','compactness_mean') ## displaying relationship between smoothness and compactness mean
```

***PCA analysis***

```python
from sklearn.decomposition import PCA
```

```python
y=df.diagnosis
y_df1= pan.get_dummies(y,drop_first=True)
# Convert data into an array
X_df = df.values

# Call PCA method
# let's call n_components = 2
pca = PCA(n_components=2)
pca_2d = pca.fit_transform(X_df)
```

```python
# Plot the PCA
plot.figure(figsize = (16,11))
plot.scatter(pca_2d[:,0],pca_2d[:,1], c = y_df1,
            cmap = "Spectral", edgecolor = "None", alpha=0.5,);
plot.title('PCA Scatter Plot');

import matplotlib.patches as mpatches
rects=[]
rects.append(mpatches.Patch(color='Red', label='Benign'));
rects.append(mpatches.Patch(color='blue', label='Malignant'));
plot.legend(handles=rects);
```

```python
X = df.drop('diagnosis', axis = 1) ## keeping independent varibale on the X-axis
y = df['diagnosis'] ## keeping dependent varibale on the Y-axis
```

```python
from sklearn.model_selection import train_test_split ## importing train test spiltting
from sklearn.metrics import classification_report ## importing classification report
from sklearn.metrics import precision_score ## importing precision score
from sklearn.metrics import recall_score ## importing recall score
from sklearn.metrics import f1_score # importing performance metrics
from sklearn.metrics import accuracy_score ## importing accuracy
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0) # splitting dataset into 3:1 ratio
```

***Decision tree***

```python
from sklearn import tree ## importing tree
dc_clf=tree.DecisionTreeClassifier(random_state=0) ## decision tree classification is defined
dc_clf.fit(X_train,y_train) ## fitting the model
```

```python
y_pred=dc_clf.predict(X_test)
dc_acc=accuracy_score(y_test,y_pred)
print("Accuracy of decision tree classifier:",dc_acc)
```

```python
print("Precision Score : ",precision_score(y_test, y_pred,
                                           average='macro')) ## calculating precision score
print("Recall Score : ",precision_score(y_test, y_pred,
                                           average='micro')) ## calculating recall score
print("f1 Score : ",f1_score(y_test, y_pred,
                                           average='weighted')) ## calculating f1 score
```

```python
from sklearn.metrics import confusion_matrix ## importing confusion matrix
```

```python
confusion_matrix(y_test, y_pred) ## displaying confusion matrix
```

***Random forest***

```python
from sklearn import ensemble
rnd_clf=ensemble.RandomForestClassifier(n_estimators=50) ## random forest is defined
rnd_clf.fit(X_train,y_train) ## fitting the model
```

```python
y_pred=rnd_clf.predict(X_test) ## validating the model
rnd_acc=accuracy_score(y_test,y_pred) ## displaying accuracy score
print("Accuracy of random forest classifier:",rnd_acc) 
```

```python
print("Precision Score : ",precision_score(y_test, y_pred,
                                           average='macro')) ## calculating precision score 
print("Recall Score : ",precision_score(y_test, y_pred,
                                           average='micro')) ## calculating recall score
print("f1 Score : ",f1_score(y_test, y_pred,
                                           average='weighted')) ## calculating f1 score
```

```python
confusion_matrix(y_test, y_pred) ## displaying Confusion matrix
```

***KNN***

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()  ## KNN is defined
```

```python
knn.fit(X_train,y_train) ## fitting the model
```

```python
y_pred=knn.predict(X_test) ## testing the model to check no of predictions
knn_acc=accuracy_score(y_test,y_pred) ## predicting the model
print("Accuracy of KNN Classifier:",knn_acc)
```

```python
print("Precision Score : ",precision_score(y_test, y_pred,
                                           average='macro')) ## calculating precision score 
print("Recall Score : ",precision_score(y_test, y_pred,
                                           average='micro')) ## calculating recall score
print("f1 Score : ",f1_score(y_test, y_pred,
                                           average='weighted')) ## calculating f1 score
```

```python
comparison = pan.DataFrame({'Model': ['Decision Tree Classifier', 
                                    'KNN Classifier', 
                                  'Random Forest Classifier',
                                 ], # setting all of the models
                        'Accuracy': [dc_acc*100, knn_acc*100, #selecting the accuracy rate
                                     rnd_acc*100]})
comparison.sort_values(by='Accuracy', ascending=False) #comparison among the accuracy rate
```
